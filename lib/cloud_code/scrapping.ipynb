```python
import requests
from bs4 import BeautifulSoup

def scrape_website(url, limit=10, depth=0):
    if depth >= limit:
        return
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Process the scraped data here
    print(f"Scraping: {url}")

    # Find all links on the page and recursively scrape them
    links = soup.find_all('a', href=True)
    for link in links:
        new_url = link['href']
        if new_url.startswith('http'):
            scrape_website(new_url, limit, depth + 1)

# Example usage
url = 'https://www.pornpics.de/milf/'
scrape_website(url)
```
